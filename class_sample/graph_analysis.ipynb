```python
import matplotlib.pyplot as plt
import pandas as pd


class Graph_Analysis():

    data_y_avg = None

    def __init__(self, data_x, data_y, data_y_avg):
        self.data_x = data_x
        self.data_y = data_y
        self.data_y_avg = data_y_avg

    def moving_average_graph(self, n=int(input('  : '))):
        df = pd.DataFrame({'Time':self.data_x, 'Data':self.data_y})
        df['moving_avg'] = df['Data'].rolling(n).mean()
        df.plot.line(y='Data', x='Time')
        df.plot.line(y='moving_avg', x='Time')

    def build_graph(self):
        plt.plot(self.data_x, self.data_y)
        plt.show()

    def build_graph_average(self):
        plt.plot(self.data_x, self.moving_average())
        plt.show()


path = ('D:/software-engineering/class_sample/data.txt')
data = []
with open(path) as f:
    for line in f:
        data.append([float(x) for x in line.split()])
data = [x for l in data for x in l]
x = tuple(data[::2])
y = tuple(data[1::2])

Graph_1 = Graph_Analysis(x, y)
Graph_1.moving_average()
```
