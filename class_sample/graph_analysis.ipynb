```python
import matplotlib.pyplot as plt
import pandas as pd


class Graph_Analysis():

    def __init__(self, data_x, data_y):
        self.data_x = data_x
        self.data_y = data_y

    def moving_average(self):
        n=int(input('Введите окно осреднения: '))
        df = pd.DataFrame({'Time':self.data_x, 'Data':self.data_y})
        df['moving_avg'] = df['Data'].rolling(n).mean()
        return df['moving_avg']

    def build_graph(self):
            df = pd.DataFrame({'Time':self.data_x, 'Data':self.data_y})
            df.plot.line(y='Data', x='Time')

    def build_moving_average_graph(self):
        n=int(input('Введите окно осреднения: '))
        df = pd.DataFrame({'Time':self.data_x, 'Data':self.data_y})
        df['moving_avg'] = df['Data'].rolling(n).mean()
        df.plot.line(y='moving_avg', x='Time')



path = ('data.txt')
data = []
with open(path) as f:
    for line in f:
        data.append([float(x) for x in line.split()])
data = [x for l in data for x in l]
x = tuple(data[::2])
y = tuple(data[1::2])

Graph_1 = Graph_Analysis(x, y)
Graph_1.moving_average()
Graph_1.build_graph()
Graph_1.build_moving_average_graph()

```
